## Plot y = sin(x) for two cycles using a scatter plot. x here is in radians

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)

plt.scatter(x,y)
plt.title('Sinusoidal Signal')
plt.xlabel('radians')
plt.ylabel('Amplitute')
plt.show()
```

## Generate a noisy sine wave x(t) = sin(2π50t)+n(t), where n∼ N(0, 0.06). On top of this, plot the clean sine wave y(t). In a bottom subplot, plot the error between x(t) and y(t). Make sure the time is aligned

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0,1,10000)
y = 3*np.sin(2*5*np.pi*t) #choosen frequency as 5hz.

Noise = np.random.normal(2,0.06,y.shape) #choosen mean = 2.

y_noise = y + Noise

plt.subplot(3, 1, 1)
plt.plot(t,y)
plt.title('y(t) = sin(2π5t)')
plt.xlabel('Radians')
plt.ylabel('Amplitude')
plt.show()

plt.subplot(3, 1, 2)
plt.plot(t,y_noise)
plt.title('Noisy signal y(t)|n = sin(2π5t) + N(0,0.06)')
plt.xlabel('Radians')
plt.ylabel('Amplitude')
plt.show()

error = y_noise - y

plt.subplot(3, 1, 3)
plt.plot(t,error)
plt.title('Error: y(t)|n - y(t)')
plt.xlabel('Radians')
plt.ylabel('Amplitude')
plt.show()
```

## Simulate 10000 throws of a biased die. The die shows odd-numbered faces with double the probability of the even number of faces. Plot the histogram of the outcome. In the same plot, overlay the histogram of 10000 throws of an unbiased die.

```python
import matplotlib.pyplot as plt
import random

# For Unbiased Dice ~ Choice is uniform
outcomes = [1,2,3,4,5,6]

# Pobabilities for Biased dice; Odd has twice prob than even
p_one = 0.222
p_two = 0.111
p_three = 0.222
p_four = 0.111
p_five = 0.222
p_six = 0.111

no_of_experiment = 100000 #Repeating the experiment 1lakh times
unbiased_occurence = []
biased_occurence = []

def rollbiaseddie():
    x = random.uniform(0,1)
    if x<=0.666:
        x = random.uniform(0,1)
        if x<0.333 and x>0: return 1
        elif x>0.333 and x<0.666: return 3
        else: return 5
    else:
        x = random.uniform(0,1)
        if x<0.333 and x>0: return 2
        elif x>0.333 and x<0.666: return 4
        else: return 6

def runexperiment():
    unbiased_occurence.append(random.choice(outcomes))
    biased_occurence.append(rollbiaseddie())

for i in range(no_of_experiment):
    runexperiment()

plt.hist(unbiased_occurence, bins = 12)
plt.title("Histogram of Unbiased Die")
plt.show()

plt.hist(biased_occurence, bins = 12)
plt.title("Histogram of Biased Die")
plt.show()
```

## Two unbiased die are thrown N times. Display the outcomes of this as a scatter plot. The size of the marker should be proportional to the count. Repeat, in a new plot for two biased dies. The plots must be self-explanatory

```python
# No of times the event to be repeated
N = 10000 

# Sample space of the event
outcomes = [1,2,3,4,5,6] 

unbiased_2dice = []

# This function is used to Get the outcome of an event
def runexperiment():
    unbiased_2dice.append([random.choice(outcomes),random.choice(outcomes)])
 

# Runs the events N times
for i in range(N):
    runexperiment()


# This calculates the no of times the outcome occured
counts = []
for i in range(1,7):
    for j in range(1,7):
        counts.append(unbiased_2dice.count([i,j]))


# Plotting the scatter plot
k = 0
for i in range(1,7):
    for j in range(1,7):
        plt.scatter(i,j, s=counts[k])
        k+=1

plt.title("Un Biased Dice")
plt.show()
```
